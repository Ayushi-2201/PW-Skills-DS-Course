```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

np.random.seed(0)
n_samples_cat = 200
categories_dist = np.random.choice(['Alpha', 'Beta', 'Gamma'], size=n_samples_cat)

numerical_dist = np.random.randn(n_samples_cat) * 5 # Base noise
numerical_dist[categories_dist == 'Alpha'] += 10 # Shift Alpha's mean
numerical_dist[categories_dist == 'Beta'] -= 5  # Shift Beta's mean
numerical_dist[categories_dist == 'Gamma'] += 2  # Shift Gamma's mean slightly

df_dist_cat = pd.DataFrame({'Category': categories_dist, 'NumericalValue': numerical_dist})

plt.figure(figsize=(9, 6))

sns.violinplot(data=df_dist_cat, x='Category', y='NumericalValue', palette='muted', inner='quartile')
# inner='quartile' shows the quartiles inside the violins
# Other options for inner: 'box', 'stick', 'point', None

plt.title("Seaborn Violin Plot: Distribution per Category")
plt.xlabel("Category")
plt.ylabel("Numerical Value Distribution")

plt.show()


plt.figure(figsize=(9, 6))
sns.boxplot(data=df_dist_cat, x='Category', y='NumericalValue', palette='muted')
plt.title("Seaborn Box Plot: Distribution per Category")
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(0)
size = 500

feat1 = np.random.rand(size) * 10
# Feature 2: Correlated with Feature 1
feat2 = 0.7 * feat1 + np.random.randn(size) * 1.5 # Positive correlation
# Feature 3: Weakly negatively correlated with Feature 1
feat3 = -0.3 * feat1 + np.random.randn(size) * 3 + 20
# Feature 4: Less correlated with others
feat4 = np.random.rand(size) * 5 + np.random.randn(size) * 2

df_corr = pd.DataFrame({
    'Feature1': feat1,
    'Feature2': feat2,
    'Feature3': feat3,
    'Feature4': feat4
})

correlation_matrix = df_corr.corr()
print(correlation_matrix) # view the matrix values

plt.figure(figsize=(8, 6))


sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
# annot=True: displays the correlation values on the heatmap
# cmap='coolwarm': sets the color map (blue=negative, red=positive)
# fmt=".2f": formats the annotation numbers to 2 decimal places
# linewidths: adds lines between cells

plt.title("Seaborn Heatmap of Feature Correlation Matrix")

plt.show()
```

```python
# 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes'
# and 'colors' columns.

np.random.seed(60)
n_points_bokeh = 100
data_scatter_bokeh = {
    'x': np.random.rand(n_points_bokeh) * 10,
    'y': np.random.rand(n_points_bokeh) * 10,
    'sizes': np.random.rand(n_points_bokeh) * 15 + 5, # Random sizes (minimum 5)
    'color_values': np.random.rand(n_points_bokeh) # Values to map to colors
}
df_scatter_bokeh = pd.DataFrame(data_scatter_bokeh)

# Create a ColumnDataSource (Bokeh's preferred data structure)
source = ColumnDataSource(df_scatter_bokeh)

# Define a color mapper
# Use 'color_values' column, map its range to the Viridis256 palette
mapper = linear_cmap(field_name='color_values', palette=Viridis256,
                     low=df_scatter_bokeh['color_values'].min(),
                     high=df_scatter_bokeh['color_values'].max())


p = figure(title="Bokeh Scatter Plot with Variable Size & Color",
           x_axis_label='Random X', y_axis_label='Random Y',
           tools="pan,wheel_zoom,box_zoom,reset,hover,save", # Add hover tool
           tooltips=[("X", "@x"), ("Y", "@y"), ("Size", "@sizes"), ("Color Value", "@color_values")]) # Configure hover tooltips

# Add a scatter renderer
p.scatter(x='x', y='y', source=source,
          size='sizes',        # Map size to 'sizes' column in source
          color=mapper,        # Map color using the mapper defined above
          legend_label="Random Points",
          alpha=0.6)           # Set transparency

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0), title="Color Value")
p.add_layout(color_bar, 'right')

show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts, color=Category10[len(fruits)]))
# Added colors directly to source for simplicity here

# Create a new plot with categorical x-axis
p = figure(x_range=fruits, height=350, title="Bokeh Bar Chart of Fruit Counts",
           x_axis_label="Fruit Type", y_axis_label="Count",
           tools="pan,wheel_zoom,box_zoom,reset,save,hover",
           tooltips=[("Fruit", "@fruits"), ("Count", "@counts")])

# Add vertical bar renderer
p.vbar(x='fruits', top='counts', width=0.8, source=source, legend_label="Fruits",
       line_color='white', fill_color='color') # Use 'color' column from source

p.xgrid.grid_line_color = None # Remove vertical grid lines
p.yaxis.axis_label_text_font_style = "bold"
p.xaxis.axis_label_text_font_style = "bold"
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.

# Given data generation
np.random.seed(70)
data_hist_bokeh = np.random.randn(1000)

# Calculate histogram bins and counts using NumPy
hist, edges = np.histogram(data_hist_bokeh, bins=30)
# hist: counts in each bin
# edges: bin edges (length is len(hist) + 1)

# Create DataFrame for ColumnDataSource (optional but often convenient)
hist_df = pd.DataFrame({'counts': hist,
                       'left': edges[:-1], # Left edge of each bin
                       'right': edges[1:]}) # Right edge of each bin
hist_df['interval'] = [f"{l:.2f} to {r:.2f}" for l, r in zip(hist_df['left'], hist_df['right'])] # Tooltip text

source_hist = ColumnDataSource(hist_df)

# Create a new plot
p = figure(height=300, width=600, title="Bokeh Histogram of Random Data",
           x_axis_label="Value Range", y_axis_label="Frequency",
           tools="pan,wheel_zoom,box_zoom,reset,save,hover",
           tooltips=[('Interval', '@interval'), ('Frequency', '@counts')]) # Hover tool shows interval and count

# Add quad glyph renderer for histogram bars
p.quad(top='counts', bottom=0, left='left', right='right', source=source_hist,
       fill_color="dodgerblue", line_color="white", alpha=0.7, legend_label="Frequency")

# Customize appearance
p.y_range.start = 0 # Ensure y-axis starts at 0
p.legend.location = "center_right"

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.

# Given data generation
np.random.seed(80)
data_heatmap_bokeh = np.random.rand(10, 10) # 10x10 matrix of random values

# --- Data Preparation for Bokeh Heatmap (using image glyph) ---
# The 'image' glyph is often easiest for matrix data. It takes the matrix directly.
# We still need coordinates for axes if we don't want simple integer indices.
# Using the provided x, y linspace:
x_hm = np.linspace(0, 1, 10) # Corresponds to columns
y_hm = np.linspace(0, 1, 10) # Corresponds to rows

# (Optional) If using x/y ranges other than matrix indices, specify them in figure()
# For image, x_range and y_range typically define the extent of the *entire* image.
# The 'x', 'y', 'dw', 'dh' parameters of image() define its position and dimensions.

# Define a color mapper
color_mapper = LinearColorMapper(palette=Plasma256, low=np.min(data_heatmap_bokeh), high=np.max(data_heatmap_bokeh))

# Create a new plot
# Let's use the actual data range for the axes based on x_hm, y_hm
x_range_hm = (x_hm[0], x_hm[-1]) # Note: image dw/dh might need adjustment depending on exact range needs
y_range_hm = (y_hm[0], y_hm[-1])

p = figure(title="Bokeh Heatmap from Matrix Data",
           x_axis_label="X Coordinate", y_axis_label="Y Coordinate",
           # Use x_range and y_range derived from x_hm, y_hm or keep as default indices (0, N-1)
           x_range=x_range_hm, y_range=y_range_hm,
           height=400, width=450, # Adjusted size for potentially square matrix
           tools="pan,wheel_zoom,box_zoom,reset,hover,save",
           tooltips=[("x", "$x"), ("y", "$y"), ("Value", "@image")]) # Tooltips for image glyph

# Add the image glyph renderer
# Note: image data expects rows -> y, columns -> x. Data might need transpose depending on interpretation.
# Assuming data_heatmap_bokeh[row_idx, col_idx] corresponds to y_hm[row_idx], x_hm[col_idx]
# The 'image' data expects a list of 2D arrays.
p.image(image=[data_heatmap_bokeh],      # The data matrix (must be in a list)
        x=x_hm[0],                       # x-coordinate of lower-left corner
        y=y_hm[0],                       # y-coordinate of lower-left corner
        dw=(x_hm[-1] - x_hm[0]),         # Width of the image in data-space units
        dh=(y_hm[-1] - y_hm[0]),         # Height of the image in data-space units
        color_mapper=color_mapper,
        level="image")

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, title="Value",
                       location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
